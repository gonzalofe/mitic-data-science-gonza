Parte V: Exploración y Visualización de Datos Avanzada

En esta quinta parte del proyecto, vamos a expandir nuestro análisis exploratorio de datos (EDA) y visualización utilizando técnicas avanzadas. Aplicaremos todo lo que hemos aprendido hasta ahora, incluyendo cuadrículas, leyendas, anotaciones, flechas, y subplots, para obtener una comprensión más profunda de nuestro conjunto de datos de ventas.

```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
df = pd.read_csv('../../../../data/retail_sales_dataset.csv')
```

```python
df.info()
```

```python
# Seleccionar solo las columnas numéricas
df_numerico = df.select_dtypes(include=[np.number])

# Calcular la matriz de correlación
correlation_matrix = df_numerico.corr()


plt.figure(figsize=(10, 8))
sns.heatmap(correlation_matrix, annot=True, cmap='coolwarm', linewidths=0.5)
plt.title('Matriz de Correlación')
plt.show()

```

Crea subplots para comparar diferentes variables clave.

Añade cuadrículas, leyendas, anotaciones y flechas a tus gráficos para mejorar la claridad y la información.

```python
import matplotlib.pyplot as plt

# Crear una figura y subplots
fig, axs = plt.subplots(1, 3, figsize=(15, 5))

# Histograma de Total Amount
axs[0].hist(df['Total Amount'], bins=20, color='blue', alpha=0.7)
axs[0].set_title('Histograma de Total Amount')
axs[0].set_xlabel('Total Amount')
axs[0].set_ylabel('Frecuencia')
axs[0].grid(True)  # Añadir cuadrícula
axs[0].legend(['Total Amount'], loc='upper right')  # Añadir leyenda
axs[0].annotate('Mayor frecuencia', xy=(50, 500), xytext=(500, 400),  # Añadir anotación y flecha
                arrowprops=dict(facecolor='black', shrink=0.05))

# Histograma de Quantity
axs[1].hist(df['Quantity'], bins=20, color='green', alpha=0.7)
axs[1].set_title('Histograma de Quantity')
axs[1].set_xlabel('Quantity')
axs[1].set_ylabel('Frecuencia')
axs[1].grid(True)  # Añadir cuadrícula
axs[1].legend(['Quantity'], loc='upper right')  # Añadir leyenda
axs[1].annotate('Distribución sesgada', xy=(3, 100), xytext=(6, 80),  # Añadir anotación y flecha
                arrowprops=dict(facecolor='black', shrink=0.05))

# Histograma de Price per Unit
axs[2].hist(df['Price per Unit'], bins=20, color='red', alpha=0.7)
axs[2].set_title('Histograma de Price per Unit')
axs[2].set_xlabel('Price per Unit')
axs[2].set_ylabel('Frecuencia')
axs[2].grid(True)  # Añadir cuadrícula
axs[2].legend(['Price per Unit'], loc='upper right')  # Añadir leyenda
axs[2].annotate('Alto precio', xy=(400, 10), xytext=(200, 50),  # Añadir anotación y flecha
                arrowprops=dict(facecolor='black', shrink=0.05))

# Ajustar el espacio entre subplots
plt.tight_layout()
plt.show()
```

Las variables "Total Amount" y "Price per Unit" parecen estar correlacionadas, ya que tanto el monto total de las ventas como el precio por unidad presentan distribuciones donde la mayor parte de las ventas son de bajo valor.
