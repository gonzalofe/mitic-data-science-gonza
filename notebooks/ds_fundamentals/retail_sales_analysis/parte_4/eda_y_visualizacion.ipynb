### Análisis Descriptivo y Visualización
Calcula estadísticas descriptivas básicas para comprender mejor las características de los datos.

```python
import pandas as pd
df = pd.read_csv(r'C:\Users\Gonzalo\OneDrive\Documentos\repos\mitic-data-science-gonza\data\retail_sales_dataset.csv')
df.describe()
```

Utiliza histogramas y boxplots para visualizar la distribución de las variables numéricas clave.

```python
import matplotlib.pyplot as plt

# Histograma de las ventas
plt.figure(figsize=(10, 6))
plt.hist(df['Total Amount'], bins=30, edgecolor='black', alpha=0.7)
plt.title('Distribución de Ventas')
plt.xlabel('Monto Total de Venta')
plt.ylabel('Frecuencia')
plt.show()

# Boxplot de las ventas
plt.figure(figsize=(10, 6))
plt.boxplot(df['Total Amount'], vert=False)
plt.title('Boxplot de Ventas')
plt.xlabel('Monto Total de Venta')
plt.show()
```

Crea gráficos de líneas para mostrar las tendencias de ventas a lo largo del tiempo.

```python
# Asegurarnos de que la columna de fecha esté en formato datetime
df['Date'] = pd.to_datetime(df['Date'])

# Agrupar por fecha para calcular el total de ventas por día
ventas_por_dia = df.groupby('Date')['Total Amount'].sum()

# Gráfico de líneas para las ventas a lo largo del tiempo
plt.figure(figsize=(12, 6))
plt.plot(ventas_por_dia.index, ventas_por_dia.values, marker='o', linestyle='-', color='b')
plt.title('Tendencia de Ventas Diarias')
plt.xlabel('Fecha')
plt.ylabel('Total de Ventas')
plt.grid(True)
plt.show()
```

Utiliza gráficos de dispersión para analizar la relación entre diferentes variables.

```python
# Gráfico de dispersión de cantidad vs. total de ventas
plt.figure(figsize=(10, 6))
plt.scatter(df['Quantity'], df['Total Amount'], alpha=0.7, edgecolor='black')
plt.title('Relación entre Cantidad de Productos Vendidos y Total de Ventas')
plt.xlabel('Cantidad de Productos Vendidos')
plt.ylabel('Monto Total de Venta')
plt.grid(True)
plt.show()
```

Combina histogramas y boxplots para visualizar la distribución de las ventas y detectar valores atípicos.

```python
# Crear una figura con dos subgráficos (histograma y boxplot)
fig, ax = plt.subplots(2, 1, figsize=(10, 8), gridspec_kw={'height_ratios': [3, 1]})

# Histograma en el primer subgráfico
ax[0].hist(df['Total Amount'], bins=30, edgecolor='black', alpha=0.7)
ax[0].set_title('Distribución de Ventas')
ax[0].set_xlabel('Monto Total de Venta')
ax[0].set_ylabel('Frecuencia')

# Boxplot en el segundo subgráfico
ax[1].boxplot(df['Total Amount'], vert=False)
ax[1].set_title('Boxplot de Ventas')
ax[1].set_xlabel('Monto Total de Venta')

# Mostrar el gráfico combinado
plt.tight_layout()
plt.show()
```
