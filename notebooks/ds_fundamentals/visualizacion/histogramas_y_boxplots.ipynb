### Crear Histogramas:
Crea histogramas para las variables Score, GDP per capita, Healthy life expectancy y Freedom to make life choices.

```python
import matplotlib.pyplot as plt
plt.hist(df['Score'], bins=5, color='skyblue', edgecolor='black')
plt.title("Distribución Puntaje de felicidad")
plt.xlabel("Puntaje")
plt.ylabel("Frecuencia")
plt.show()
```

```python
plt.hist(df['GDP per capita'], bins=5, color='skyblue', edgecolor='black')
plt.title("Distribución de PIB per capita")
plt.xlabel("PIB per capita")
plt.ylabel("Frecuencia")
plt.show()
```

```python
plt.hist(df['Healthy life expectancy'], bins=5, color='skyblue', edgecolor='black')
plt.title("Distribución de Expectancia de vida saludable")
plt.xlabel("Expectancia de vida saludable")
plt.ylabel("Frecuencia")
plt.show()
```

```python
plt.hist(df['Freedom to make life choices'], bins=5, color='skyblue', edgecolor='black')
plt.title("Distribución de libertad para tomar decisiones")
plt.xlabel("Libertad para tomar decisiones")
plt.ylabel("Frecuencia")
plt.show()
```

### Crear Boxplots:
Crea boxplots para las mismas variables para visualizar la mediana, el rango intercuartil y los valores atípicos. Personaliza los boxplots con diferentes colores y estilos.

```python
plt.boxplot(df['Score'], notch=True, patch_artist=True, boxprops=dict(facecolor='skyblue', color='black'), medianprops=dict(color='red'), whiskerprops=dict(color='black'), capprops=dict(color='black'))
plt.title("Boxplot de puntaje")
plt.ylabel("Puntaje")
plt.grid(True)
plt.show()
```

```python
plt.boxplot(df['GDP per capita'], notch=True, patch_artist=True, boxprops=dict(facecolor='blue', color='yellow'), medianprops=dict(color='green'), whiskerprops=dict(color='red'), capprops=dict(color='blue'))
plt.title("Boxplot de GDP per capita")
plt.ylabel("GDP per capita")
plt.grid(True)
plt.show()
```

```python
plt.boxplot(df['Healthy life expectancy'], notch=True, patch_artist=True, boxprops=dict(facecolor='green', color='yellow'), medianprops=dict(color='black'), whiskerprops=dict(color='red'), capprops=dict(color='blue'))
plt.title("Boxplot de expectancia de vida saludable")
plt.ylabel("Expectancia de vida saludables")
plt.grid(True)
plt.show()
```

```python
plt.boxplot(df['Freedom to make life choices'], notch=True, patch_artist=True, boxprops=dict(facecolor='green', color='yellow'), medianprops=dict(color='black'), whiskerprops=dict(color='red'), capprops=dict(color='blue'))
plt.title("Boxplot de libertad para tomar decisiones")
plt.ylabel("Libertad para tomar decisiones")
plt.grid(True)
plt.show()
```

### Análisis Combinado:
Combina los histogramas y los boxplots en una sola figura para cada variable.

```python

variables = ['Score', 'GDP per capita', 'Healthy life expectancy', 'Freedom to make life choices']

# Creating 4 separate figures, each containing a histogram and boxplot for a variable

for var in variables:
    fig, axs = plt.subplots(1, 2, figsize=(10, 5))

    # Histogram
    axs[0].hist(df[var], bins=20, color='lightblue', edgecolor='black', alpha=0.7)
    axs[0].set_title(f'Histogram of {var}')
    axs[0].set_xlabel(var)
    axs[0].set_ylabel('Frequency')

    # Boxplot
    axs[1].boxplot(df[var], patch_artist=True, boxprops=dict(facecolor='lightcoral', color='black'),
                   medianprops=dict(color='blue'), whiskerprops=dict(color='black'))
    axs[1].set_title(f'Boxplot of {var}')
    axs[1].set_ylabel(var)

    # Adjust layout and show
    plt.tight_layout()
    plt.show()
```

1. Score:
Distribución: El histograma muestra una distribución más o menos normal, con la mayoría de los países teniendo un puntaje entre 4 y 7. La distribución está levemente sesgada hacia la derecha, lo que indica que hay más países con puntajes altos que bajos.
Boxplot: El boxplot revela una distribución relativamente simétrica, con algunos valores atípicos en el extremo inferior. Esto significa que hay unos pocos países con puntajes notablemente bajos, pero en general la mayoría de los puntajes están concentrados cerca de la mediana.
Valores atípicos: Algunos países con puntajes bajos se destacan como valores atípicos, pero no son extremadamente distantes de la mayoría.
2. GDP per capita:
Distribución: El histograma muestra una fuerte asimetría hacia la derecha, con muchos países concentrados en el rango inferior de PIB per cápita y unos pocos países con valores altos que extienden la cola de la distribución. Esto es común en datos de riqueza, donde la mayoría de los países tienen un PIB más bajo y unos pocos tienen ingresos extremadamente altos.
Boxplot: El boxplot refuerza esta interpretación, mostrando que la mayor parte de los datos están agrupados en el rango bajo, mientras que hay varios valores atípicos hacia el extremo superior.
Valores atípicos: Los valores atípicos representan países con un PIB per cápita mucho más alto que la mediana global, lo cual es esperable considerando las grandes diferencias económicas entre países.
3. Healthy life expectancy:
Distribución: El histograma muestra una distribución relativamente compacta y simétrica, con la mayoría de los países teniendo una expectativa de vida saludable entre 0.8 y 1.2. No hay una fuerte asimetría en los datos.
Boxplot: El boxplot refleja esta distribución ajustada, con una mediana claramente definida y un rango intercuartil relativamente pequeño. No hay muchos valores atípicos.
Valores atípicos: Solo se observan algunos valores atípicos en los rangos más bajos, lo que podría indicar países con problemas serios en su sistema de salud o condiciones de vida difíciles que afectan la expectativa de vida.
4. Freedom to make life choices:
Distribución: El histograma muestra una distribución más dispersa, con valores que van desde alrededor de 0.3 hasta más de 0.7. La mayoría de los países están agrupados en el rango medio, pero hay una distribución más amplia que en otras variables.
Boxplot: El boxplot muestra una mayor dispersión en comparación con las otras variables. El rango intercuartil es más amplio y hay valores atípicos tanto en los rangos altos como bajos.
Valores atípicos: Los valores atípicos en ambos extremos indican que algunos países tienen niveles excepcionalmente altos de libertad para tomar decisiones de vida, mientras que otros tienen niveles muy bajos. Esto refleja la variabilidad en las políticas y derechos individuales entre países.
Conclusión General:
Las variables Score y Healthy life expectancy muestran distribuciones más compactas y simétricas, con pocos valores atípicos.
GDP per capita y Freedom to make life choices tienen distribuciones más asimétricas y presentan más valores atípicos, lo que sugiere una mayor variabilidad en estas medidas entre los países.
Los valores atípicos en el PIB y la libertad de elección son esperables debido a las grandes disparidades económicas y sociales entre los países representados en el dataset.
